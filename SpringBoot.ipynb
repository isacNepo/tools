## Salvar o projeto no Github

1. Abrir terminal do GitBash na pasta do projeto

2. `git init`

3. `git add .`

4. `git commit -m "Backend project created"`

5. Criar um novo projeto no GitHub

6. `git remote add origin (URL do git do projeto)`

7. `git push -u origin main`
