#### String e Indexação

````python
# Atribuindo uma string
palavra = 'Data Science Academy'
````

````python
print(palavra)
````

````python
# Primeiro elemento da string. 
palavra[0]
````

````python
palavra[1]
````

````python
palavra[2]
````

Podemos usar um : para executar um slicing que faz a leitura de tudo até um ponto designado. Por exemplo:

````python
# Retorna todos os elementos da string, começando pela posição (lembre-se que Python começa a indexação pela posição 0),
# até o fim da string.
palavra[1:]
````

````python
# A string original permanece inalterada, pois é imutavel
palavra
````

````python
# Retorna tudo até a posição 3
palavra[:3]
````

````python
palavra[:]
````

````python
# Nós também podemos usar a indexação negativa e ler de trás para frente.
palavra[-1]
````

````python
# Retornar tudo, exceto a última letra
palavra[:-1]
````

Nós também podemos usar a notação de índice e fatiar a string em pedaços específicos (o padrão é 1). Por exemplo, podemos usar dois pontos duas vezes em uma linha e, em seguida, um número que especifica a frequência para retornar elementos. Por exemplo:

````python
palavra[::1]
````

````python
palavra[::2]
````

````python
palavra[::-1]
````

### Dicionários

Os dicionários são construidos com o uso de chaves { } e virgulas separando cada elementos do dicionario

Exemplo: `dict = {k1:v1, k2:v2, ..., kn:vn}`

O que são mapeamentos?

Mapeamentos são uma coleção de objetos que são armazenados por uma chave, ao contrario de uma sequncia de objetos armazenados por sua posição relativa.

````python
# Isso é uma lista
estudantes_lst = ["Mateus", 24, "Fernanda", 22, "Tamires", 26, "Cristiano", 25] 
````

````python
estudantes_lst
````

````python
# Isso é um dicionário
estudantes_dict = {"Mateus":24, "Fernanda":22, "Tamires":26, "Cristiano":25}
````

````python
estudantes_dict 
````

````python
# Ver qual o valor (chave) de "Matheus"
estudantes_dict["Mateus"]
````

````python
# Atribui um novo item ao dicionario
estudantes_dict["Pedro"] = 23
````

````python
# Limpa o dicionario
estudantes_dict.clear() 
````

````python
estudantes_dict
````

````python
# Deleta o dicionario 
del estudantes_dict
````

````python
estudantes_dict
````

````python
estudantes = {"Mateus":24, "Fernanda":22, "Tamires":26, "Cristiano":25}
````

````python
estudantes
````

````python
# Montra o tamanho do dicionario
len(estudantes)
````

````python
# Monstra apenas as chaves
estudantes.keys()
````

````python
# Monstra apenas os valores
estudantes.values()
````

````python
# Montra todos os itens de forma separada (chave/valor)
estudantes.items()
````

````python
estudantes2 = {"Maria":27, "Erika":28, "Milton":26}

# Atualiza os dados do dicionario "estudantes"
estudantes.update(estudantes2)

estudantes
````

### Tubla

Tublas são semelhantes as listas, porem são imutaveis. Por isso utiliza-se tublas para apresentar dados que não podem ser alterados, como dias da semana ou datas em um calendario.

````python
# Criando uma tupla
tupla1 = ("Geografia", 23, "Elefantes")
````

````python
# Imprimindo a tupla
tupla1
````

````python
# Tuplas não suportam append()
tupla1.append("Chocolate")   
````

````python
# Tuplas não suportam delete de um item específico
del tupla1["Gatos"]  
````

````python
# Tuplas não suportam atribuição de item
tupla1[1] = 21
````

````python
# Verificando o comprimento da tupla
len(tupla1)
````

````python
# Deletando a tupla
del tupla1
````

````python
tupla1
````

````python
# Criando uma tupla
t2 = ('A', 'B', 'C')

# Usando a função list() para converter uma tupla para lista
lista_t2 = list(t2)

lista_t2
````

````python
lista_t2.append('D')
````

````python
# Usando a função tuple() para converter uma lista para tupla
t2 = tuple(lista_t2)

t2
````
